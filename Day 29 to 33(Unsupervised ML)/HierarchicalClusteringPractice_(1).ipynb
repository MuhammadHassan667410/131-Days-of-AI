```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'/content/internet_usage.csv')
df.head()
```

```python
df1 = df.drop(df[['Country Name', 'Country Code']], axis = 1)
```

```python
df1.replace('..', np.nan, inplace=True)
```

```python
df1
```

```python
for col in df1.columns:
    df1[col] = df1[col].astype('float64')
```

```python
df1.info()
```

```python
for col in df1.columns:
    df1[col].fillna(df1[col].mean(), inplace=True)
```

```python
df1
```

```python
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(10,7))
plt.title('Dendogram', fontweight='bold')
dend = shc.dendrogram(shc.linkage(df1, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3)
labels_ = model.fit_predict(df1)
```

```python
labels_
```

```python
import seaborn as sns

sns.scatterplot(x=df1.iloc[:, 22], y=df1.iloc[:, 23], hue=model.labels_)
plt.title("Agglomerative Clustering")
plt.xlabel('2022')
plt.ylabel('2023')
```
